```python
from dotenv import load_dotenv
import os
from back import LLMHandler, VectorDatabase, QuestionAnsweringChain
```

```python
#load llm
load_dotenv()
gemini_key = os.getenv('gemini_key')
qdrant_key=os.getenv('qdrant_key')
# Initialize the classes
llm_handler = LLMHandler(model_name="gemini-1.5-flash", gemini_key=gemini_key)
vector_db = VectorDatabase(
        model_name="hiieu/halong_embedding",
        collection_name='cmc_final_db',
        api=qdrant_key
    )
```

```python
qa_chain = QuestionAnsweringChain(llm_handler, vector_db,num_docs=5, apply_rerank=True, apply_rewrite=True, date_impact= 0.001)
```

```python
with open('question_eval.txt','r', encoding='utf-8') as file:
    questions=[x for x in file.read().split('\n') if len(x)>10]
answers=[]
with open('responses.txt', 'w', encoding='utf-8') as f:
    for i in questions:
        response,links=qa_chain.run(i)
        f.write(response+'\n'+','.join(links)+'\n\n')
    
```

```python
questions = [
    "CMC là gì và khi nào được thành lập?",
    "CMC cung cấp những dịch vụ nào?",
    "Địa chỉ trụ sở chính của CMC là ở đâu?",
    "CMC có các giải pháp phần mềm nào cho doanh nghiệp?",
    "Ai là Giám đốc điều hành của CMC hiện tại?",
    "CMC có hợp tác gì với các doanh nghiệp hàn quốc",
    "CMC có hợp tác với những đối tác lớn nào?",
    "Danh sách các khách hàng tiêu biểu của CMC là ai?",
    "CMC có chính sách bảo mật thông tin khách hàng như thế nào?",
    "Các điều khoản dịch vụ của CMC có thể tìm thấy ở đâu?",
    "Làm thế nào tôi có thể liên hệ với bộ phận hỗ trợ khách hàng của CMC?",
    "CMC cung cấp dịch vụ hỗ trợ khách hàng vào thời gian nào?",
    "CMC hiện đang tuyển dụng những vị trí nào?",
    "Tôi có thể tìm thông tin về cơ hội việc làm tại CMC ở đâu?",
    "CMC có các dự án nghiên cứu và phát triển nào đang hoạt động?",
    "Làm thế nào CMC đảm bảo sự đổi mới trong các sản phẩm của mình?",
    "CMC có công bố báo cáo tài chính hàng năm không?",
]

answers=[]
for ques in questions:
    qa_chain.memory=[]
    # Run the chain with a question
    result = qa_chain.run(ques)
    answers.append(result)
    print('response: \n'+result[0])
    print('+'*100)
# Save to a text file
with open('responses.txt', 'w', encoding='utf-8') as f:
    for idx,response in enumerate(answers):
        f.write('\nquestion:\n'+questions[idx]+'\nanswer:\n'+response[0] + '\nsource:\n'+response[1])
```

```python
multi_turn_tests = [

    {
        "conversation": [
            "CMC cung cấp những sản phẩm nào?",  # Initial query about products.
            "giới thiệu về sản phẩm đầu tiên trong những sản phẩm trên ",  # Narrow down to software products.
            "nó có gì đặc biệt"  # Follow-up on key products.
        ]
    },


    {
        "conversation": [
            "CMC hợp tác với những đối tác nào?",  # General question about partners.
            "Đối tác lớn nhất của họ là ai?",  # Narrow down to key partners.
            "Công ty có hợp tác với các đối tác quốc tế không?"  # Ask about international partnerships.
        ]
    },

    {
        "conversation": [
            "CMC có những dự án R&D nào?",  # General question on R&D.
            "Những dự án này tập trung vào lĩnh vực nào?",  # Follow-up on R&D focus areas.
            "Dự án nổi bật nhất hiện tại là gì?"  # Ask about the most prominent project.
        ]
    },

]
multi_answer=[]
for conversation in multi_turn_tests:
    temp=[]
    qa_chain.memory=[]
    for questi in conversation['conversation']:
        answer=qa_chain.run(questi)
        temp.append(answer)
        print('\nresponse: \n'+answer[0])
        print('-'*100)
    multi_answer.append(temp)
```

```python
with open('conversation.txt', 'w', encoding='utf-8') as f:
    for i,conver in enumerate(multi_turn_tests): 
        f.write(f'\nconversation {i}\n')
        for idx,response in enumerate(multi_answer[i]):
            f.write('\nquestion:\n'+multi_turn_tests[i]['conversation'][idx]+'\nanswer:\n'+response[0] + '\nsource:\n'+response[1])
```

```python
compound_queries = [
    "CMC có cung cấp dịch vụ phần mềm và hợp tác với các đối tác quốc tế nào không?",
    "Ai là CEO của CMC và công ty này có những sản phẩm công nghệ nào nổi bật?",
    "CMC có những chính sách bảo mật nào cho khách hàng và những đối tác chính của họ là ai?",
    "CMC thành lập năm nào và hiện tại công ty có bao nhiêu nhân viên?",
    "Những dịch vụ tư vấn của CMC là gì và các sự kiện lớn gần đây mà công ty đã tổ chức là gì?"
]
combound_answer=[]
for ques in compound_queries:
    qa_chain.memory=[]
    # Run the chain with a question
    result = qa_chain.run(ques)
    combound_answer.append(result)
    print('response: \n'+result[0])
    print('+'*100)
```

```python
with open('compound_query.txt', 'w', encoding='utf-8') as f:
    for idx, com_ans in enumerate(combound_answer):
        f.write('\nquestion:\n'+compound_queries[idx]+'\nanswer:\n'+com_ans[0] + '\nsource:\n'+com_ans[1])
```
