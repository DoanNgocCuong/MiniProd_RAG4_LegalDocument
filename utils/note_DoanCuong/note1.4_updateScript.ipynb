Dựa vào file `crul.md`, tôi thấy cần cập nhật lại format request trong `script.js` để khớp với API endpoint. Đây là phần cần sửa:

````javascript:deploy/frontend/script.js
async function sendMessage(message) {
    if (!message) return;

    displayMessage(message, 'user');

    try {
        const response = await fetch(`http://localhost:3000/send_message`, {
            method: 'POST',
            headers: {
                'Content-Type': 'application/json',
                'Accept': 'application/json'
            },
            credentials: 'include',
            mode: 'cors',
            body: JSON.stringify({
                question: message  // Thay đổi từ message thành question
                // Bỏ các field không cần thiết: user_id, conversation_id
            })
        });

        if (!response.ok) {
            throw new Error(`HTTP error! status: ${response.status}`);
        }

        const data = await response.json();
        // Cập nhật cách xử lý response theo format mới
        displayMessage(data.response, 'bot');  // data.response thay vì data.response.answer
        
        // Hiển thị links nếu có
        if (data.links && data.links.length > 0) {
            const linksMessage = 'Related links:\n' + data.links.join('\n');
            displayMessage(linksMessage, 'bot');
        }
    } catch (error) {
        console.error('Error sending message:', error);
        displayMessage('Sorry, there was an error processing your message.', 'bot');
    }
}
````

Những thay đổi chính:
1. Đơn giản hóa body request chỉ còn field `question`
2. Cập nhật cách xử lý response để phù hợp với format API trả về
3. Thêm phần hiển thị links nếu có
4. Bỏ phần xử lý `conversation_id` vì API không yêu cầu

Format request mới sẽ khớp với ví dụ trong `crul.md`.

```
